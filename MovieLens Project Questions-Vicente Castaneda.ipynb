### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
item_year = item.set_index('year')#Function to make year column as the index
```

```python
item_year.head()
```

```python
year_list1 = item_year
year_list1['Total Sum Per Year'] = year_list1.sum(axis=1) #function to obtain sum per year
year_list1.head(10) #check to see if formula worked
```

```python
genre_by_year= year_list1.drop('movie id',axis=1,inplace=True), year_list1.drop('movie title', axis=1, inplace=True), year_list1.drop('release date', axis=1, inplace=True), year_list1.drop('Total Sum Per Year', axis=1, inplace=True)
#function to drop columns from dataframe
```

```python
genre_by_year = year_list1.groupby('year').sum() #function to group and sum all genres by year
genre_by_year
```

```python
genre_percent = (genre_by_year.div(genre_by_year['Total'], axis= 0) * 100)# Function to divide total by 100 to obtain %
genre_percent
```

```python
plt.figure(figsize=(20, 20)) #function to increase size of dataframe

sns.heatmap(genre_percent,annot=True,cmap='plasma',vmin=-1,vmax=1) #heatmap function
```

```python
rating_data=pd.read_csv('Data.csv') #import data dataset
movie_data=pd.read_csv('item.csv') #import item dataset
user_data=pd.read_csv('user.csv') #import user dataset

info_movie_rating = pd.merge(rating_data, movie_data) #funciton to merge datasets for rating and movies
```
